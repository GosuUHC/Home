```python
%matplotlib inline
import matplotlib.pyplot as plt
import timeit
```

```python
def foo(i):  # i - число
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i = i // 10
    return result


time = timeit.Timer("foo(num)", globals=globals())

plt_x = []
plt_time = []


num = 123
for i in range(3, 2000, 1):
    plt_x.append(i)

    t = time.timeit(number=5)
    plt_time.append(t)

    num *= 10


plt.plot(plt_x, plt_time)
plt.show()
```

1. Функция преобразует число в строку
2. O(2^n)

**Дан список. Проверить, что все значения списка различны.**

```python
import random


def unique_list_to_set_len(test_lst):
    lizt = test_lst
    list_len = len(lizt)
    set_ = set(lizt)
    set_len = len(set_)
    if list_len == set_len:
        return True
    return False


def unique_classic(test_lst):
    N = len(test_lst)
    for i in range(N - 1):
        for j in range(i + 1, N):
            if test_lst[i] == test_lst[j]:
                return False
    return True


time1 = timeit.Timer("unique_list_to_set_len(lst)", globals=globals())
time2 = timeit.Timer("unique_classic(lst)", globals=globals())

plt_x = []
plt_time1 = []
plt_time2 = []

for i in range(1000, 10001, 1000):
    # из диапазона () взять i случайных чисел
    lst = random.sample(range(-100000, 100000), i)
    plt_x.append(i)

    t = time1.timeit(number=5)
    plt_time1.append(t)

    t = time2.timeit(number=5)
    plt_time2.append(t)

plt.plot(plt_x, plt_time1, plt_x, plt_time2)


plt.show()
```

2. Сложность первого O(1). Сложность второго O(n^2) 

#### **Задание 3**
Количество операций при выполнении двух алгоритмов для массива размером N таково:

![title](task3.PNG)

Определите размер массива N, для которого время выполнения обоих алгоритмов
одинаково.

```python
def first_t(n):
    return (n**2 - n - 10)


def second_t(n):
    return (4*n + 40)


n = 0
res1 = first_t(n)
res2 = second_t(n)
plt_lst1 = [res1]
plt_lst2 = [res2]
plt_x = [n]
while True:
    n += 1
    plt_x.append(n)
    res1 = first_t(n)
    plt_lst1.append(res1)
    res2 = second_t(n)
    plt_lst2.append(res2)
    if res1 == res2:
        break

print("Время будет одинаково при N =", n)

plt.plot(plt_x, plt_lst1, plt_lst2)
plt.show()
```
