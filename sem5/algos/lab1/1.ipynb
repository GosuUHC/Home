```python
%matplotlib inline
import matplotlib.pyplot as plt
import timeit
```

```python
def foo(i):  # i - число
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i = i // 10
    return result


time = timeit.Timer("foo(num)", globals=globals())

plt_x = []
plt_time = []


count = 0
num = 123
while count < 2000:
    plt_x.append(count)

    t = time.timeit(number=5)
    plt_time.append(t)
    count += 1
    num *= 10

plt.plot(plt_x, plt_time)
plt.show()
```

1. Функция преобразует число в строку
2. O(2^n)

#### **Задание 3**
Количество операций при выполнении двух алгоритмов для массива размером N таково:

![title](task3.PNG)

Определите размер массива N, для которого время выполнения обоих алгоритмов
одинаково.

```python
def first_t(n):
    return (n**2 - n - 10)


def second_t(n):
    return (4*n + 40)


n = 0
res1 = first_t(n)
res2 = second_t(n)
plt_lst1 = [res1]
plt_lst2 = [res2]
plt_x = [n]
while True:
    n += 1
    plt_x.append(n)
    res1 = first_t(n)
    plt_lst1.append(res1)
    res2 = second_t(n)
    plt_lst2.append(res2)
    if res1 == res2:
        break

print("Время будет одинаково при N =", n)

plt.plot(plt_x, plt_lst1, plt_lst2)
plt.show()
```
