```python
%matplotlib inline
import matplotlib.pyplot as plt
import timeit
```

```python
def foo(i):  # i - число
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i = i // 10
    return result

time = timeit.Timer("foo(i)", globals=globals())

plt_x = []
plt_time = []

for i in range(100, 100000, 2):
    plt_x.append(i)

    t = time.timeit(number=100)
    plt_time.append(t)

plt.plot(plt_x, plt_time)
plt.show()
```

**Дан список. Проверить, что все значения списка различны.**

```python
test_list1 = list(i for i in range(1, 10000, 1))
test_list2 = list(map(lambda x: x - 1 if x % 2 == 0 else x, test_list1))


def unique_list_to_set_len(test_list):
    list_len = len(test_list)
    set_len = len(set(test_list))
    if list_len == set_len:
        return True
    return False


for i in test_list1, test_list2:
    if unique_list_to_set_len(i):
        print("Все элементы списка различны")
    else:
        print("Есть одинаковые элементы в списке")


def test(test_list):
    N = len(test_list)
    for i in range(N - 1):
        for j in range(i + 1, N):
            if test_list[i] == test_list[j]:
                return False
    return True


for i in test_list1, test_list2:
    if test(i):
        print("Все элементы списка различны")
    else:
        print("Есть одинаковые элементы в списке")
```

#### **Задание 3**
Количество операций при выполнении двух алгоритмов для массива размером N таково:

![title](task3.PNG)

Определите размер массива N, для которого время выполнения обоих алгоритмов
одинаково.

```python
def first_t(n):
    return (n**2 - n - 10)


def second_t(n):
    return (4*n + 40)


n = 0
res1 = first_t(n)
res2 = second_t(n)
plt_lst1 = [res1]
plt_lst2 = [res2]
plt_x = [n]
while True:
    n += 1
    plt_x.append(n)
    res1 = first_t(n)
    plt_lst1.append(res1)
    res2 = second_t(n)
    plt_lst2.append(res2)
    if res1 == res2:
        break

print("Время будет одинаково при N =", n)

plt.plot(plt_x, plt_lst1, plt_lst2)
plt.show()
```
